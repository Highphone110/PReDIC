```python
# Imports needed:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
fileame = '0002_result.csv' # The original pyDIC output
deformation = pd.read_csv(fileame)
deformation_np = np.array(deformation)
```

### File output:
The output file of pyDIC is more suited to a pandas array as the header is already built into the file.

The headers are:

index, index_x, index_y, pos_x, pos_y, disp_x, disp_y, strain_xx, strain_yy, strain_xy

```python
deformation.loc[:,'index_x':'strain_xy']
```

```python
x_ind = np.reshape(deformation_np[:,1], (26,26))
y_ind = np.reshape(deformation_np[:,2], (26,26))
disp_x = np.reshape(deformation_np[:,5], (26,26))
disp_y = np.reshape(deformation_np[:,6], (26,26))
strain_xx = np.reshape(deformation_np[:,7], (26,26))
strain_yy = np.reshape(deformation_np[:,8], (26,26))
strain_xy = np.reshape(deformation_np[:,9], (26,26))
```

```python
# Example of what the entire array looks like
imgplot = plt.imshow(np.flipud(np.rot90(disp_x)), origin="lower")
imgplot.set_cmap('YlGnBu')
plt.colorbar()
plt.show()
```

```python
# Example of what the entire array looks like
imgplot = plt.imshow(np.rot90(disp_y), origin="lower")
imgplot.set_cmap('YlGnBu')
plt.colorbar()
plt.show()
```

```python
residual = (np.flipud(np.rot90(disp_x))**2 + np.rot90(disp_y)**2 )**0.5

# Example of what the entire array looks like
imgplot = plt.imshow(residual, origin="lower")
imgplot.set_cmap('YlGnBu')
plt.colorbar()
plt.show()
```

```python
# Example of what the entire array looks like
imgplot = plt.imshow(np.rot90(strain_xx), origin="lower")
imgplot.set_cmap('nipy_spectral')
plt.colorbar()
plt.show()
```

```python
# Example of what the entire array looks like
imgplot = plt.imshow(np.rot90(strain_yy), origin="lower")
imgplot.set_cmap('nipy_spectral')
plt.colorbar()
plt.show()
```
